```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 50)
# print(x)
y = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure()
plt.plot(x, y, color="mediumspringgreen", label="sine", linewidth=10, alpha=1, linestyle="dashdot")
plt.scatter(x, y2, color="rebeccapurple", label="cosine", marker="3")
plt.plot(x, y2*2, color="lightpink", label="i added it for funsies", marker=(8,1,90))
plt.title("example")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="upper right")
plt.show()
```
