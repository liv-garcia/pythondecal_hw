```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def sine_wave ( x, A, w ):

    x = np.linspace(0, 2*np.pi, 1000)
    y = A * np. sin(w * x)
    return x, y 

A_arr = np.array([1,3,7,5,10]) 
w_arr = np.array([0.5, 0.75, 1, 1.5, 1.75])

# Loop through each pair of A and w values. For each iteration, calculate
# the array of y data that correspond to those values, using the same x data
# array for each iteration. Plot each array of y data onto your figure

x, y = sine_wave(x=0, A=1, w=1)

plt.figure()

for i in range(len(A_arr)):
    x, y = sine_wave(x=0, A= A_arr[i], w= w_arr[i]) 
    plt.plot(x,y)
    
plt.title("5 Different Sine Functions")
plt.xlabel("x data")
plt.ylabel("y data")
plt.legend(['sine function 1','sine function 2','sine function 3', 'sine function 4','sine function 5'])
plt.show()
```
